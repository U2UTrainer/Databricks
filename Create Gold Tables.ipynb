```sql
CREATE CATALOG IF NOT EXISTS dev_gold;
USE CATALOG dev_gold;
```

```sql
CREATE SCHEMA IF NOT EXISTS sales;
USE SCHEMA sales;
```

```sql
CREATE TABLE DimProduct
(
  ProductKey INTEGER NOT NULL PRIMARY KEY,
  ProductName STRING NOT NULL,
  ProdductSubcategory STRING NOT NULL,
  ProductCategory STRING NOT NULL,
  Color STRING NOT NULL,
  ListPrice DECIMAL(10,2) NOT NULL
)
```

```sql
CREATE TABLE DimCustomer
(
  CustomerKey INTEGER NOT NULL PRIMARY KEY,
  FullName STRING NOT NULL,
  Gender STRING NOT NULL,
  MaritalStatus STRING NOT NULL,
  YearlyIncome DECIMAL(10,2) NOT NULL,
  TotalChildren INTEGER NOT NULL,
  Education STRING NOT NULL,
  Occupation STRING NOT NULL,
  City STRING NOT NULL,
  StateProvince STRING NOT NULL,
  Country STRING NOT NULL
)
```

```sql
CREATE TABLE DimDate
(
  DateKey INTEGER NOT NULL PRIMARY KEY,
  TheDate DATE NOT NULL,
  MonthName STRING NOT NULL,
  MonthNumberOfYear INTEGER NOT NULL,
  CalendarQuarter STRING NOT NULL,
  CalendarYear INTEGER NOT NULL
)
```

```sql
CREATE TABLE FactSales
(
  ProductKey INTEGER NOT NULL FOREIGN KEY REFERENCES DimProduct(ProductKey),
  CustomerKey INTEGER NOT NULL FOREIGN KEY REFERENCES DimCustomer(CustomerKey),
  OrderDateKey INTEGER NOT NULL FOREIGN KEY REFERENCES DimDate(DateKey),
  OrderQuantity INTEGER NOT NULL,
  SalesAmount DECIMAL(10,2) NOT NULL,
  DiscountAmount DECIMAL(10,2) NOT NULL
)
```
