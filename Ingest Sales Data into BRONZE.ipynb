```python
%sql
USE CATALOG dev_bronze;
USE SCHEMA sales;
```

```python
%sql
DROP TABLE IF EXISTS customer;
DROP TABLE IF EXISTS geography;
DROP TABLE IF EXISTS product;
DROP TABLE IF EXISTS productsubcategory;
DROP TABLE IF EXISTS productcategory;
DROP TABLE IF EXISTS date;
DROP TABLE IF EXISTS sales;
```

```python
landingZonePath = '/Volumes/dev_bronze/sales/landingzone'
```

```python
dfCustomer = spark.read.format('csv').option('header','true').option('inferSchema','true').load(f'{landingZonePath}/customer.csv')
```

```python
dfGeography = spark.read.format('csv').option('header','true').option('inferSchema','true').load(f'{landingZonePath}/geography.csv')
```

```python

dfProduct = spark.read.format('csv').option('header','true').option('inferSchema','true').load(f'{landingZonePath}/product.csv')
```

```python
dfProductSubcategory = spark.read.format('csv').option('header','true').option('inferSchema','true').load(f'{landingZonePath}/productsubcategory.csv')
```

```python
dfProductCategory = spark.read.format('csv').option('header','true').option('inferSchema','true').load(f'{landingZonePath}/productcategory.csv')
```

```python

dfDate = spark.read.format('csv').option('header','true').option('inferSchema','true').load(f'{landingZonePath}/date.csv')
```

```python
dfInternetSales = spark.read.format('csv').option('header','true').option('inferSchema','true').load(f'{landingZonePath}/internetsales.csv')
```

```python
%sql
USE CATALOG dev_bronze;
USE SCHEMA sales;
```

```python
dfCustomer.write.saveAsTable('customer')
dfGeography.write.saveAsTable('geography')
dfProduct.write.saveAsTable('product')
dfProductSubcategory.write.saveAsTable('productsubcategory')
dfProductCategory.write.saveAsTable('productcategory')
dfDate.write.saveAsTable('date')
dfInternetSales.write.saveAsTable('sales')
```
